## Transformación de Archivos XLSX a Datasets en Python

Es totalmente válido querer simplificar tu flujo de trabajo, especialmente si estás usando un PC con bajos recursos. Convertir archivos **XLSX** a **DataFrames** utilizando **pandas** es una excelente opción. Esta librería es eficiente y gratuita. A continuación, te explico cómo hacerlo.

<hr>

### Requisitos Previos

Asegúrate de tener instalado el paquete `pandas` y `openpyxl` para manejar archivos Excel. Si no los tienes, instálalos con:

```python
import pandas as pd
```

### Cargar el archivo XLSX

1. **Leer el archivo XLSX**:
   Utiliza la función `read_excel` de pandas para cargar tu archivo.

```python
df_productos = pd.read_excel('../data/productos.xlsx')
df_clientes = pd.read_excel('../data/clientes.xlsx')
df_ventas = pd.read_excel('../data/ventas.xlsx')
df_detalle_ventas = pd.read_excel('../data/detalle_ventas.xlsx')
```

### Limpiar datos
2. **Transformar y manipular el DataFrame**:
   Aquí puedes hacer todas las transformaciones necesarias, como limpiar datos o filtrar columnas.

```python
df_productos.dropna(inplace=True)
df_clientes.dropna(inplace=True)
df_clientes.dropna(inplace=True)
df_detalle_ventas.dropna(inplace=True)
```

### Guardar en formato CSV
3. **Guardar el DataFrame**:
   Una vez que tu DataFrame esté listo, puedes guardarlo en un nuevo archivo CSV o en un formato de tu elección.

```python
df_productos.to_csv('../data/csv/origin/productos.csv', index=False)
df_clientes.to_csv('../data/csv/origin/clientes.csv', index=False)
df_ventas.to_csv('../data/csv/origin/ventas.csv', index=False)
df_detalle_ventas.to_csv('../data/csv/origin/detalle_ventas.csv', index=False)
```

Ahora puedo trabajar con dataset.csv

### Ventajas de Este Enfoque

- **Eficiencia**: Trabajar con archivos CSV es generalmente más rápido.
- **Bajos Recursos**: Los archivos CSV ocupan menos memoria.
- **Sin Costo**: Todo el proceso se realiza con herramientas gratuitas.
